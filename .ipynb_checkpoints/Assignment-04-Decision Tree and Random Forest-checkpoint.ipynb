```python
import numpy as np
import matplotlib
from matplotlib import pyplot as plt
```

```python
import sklearn.linear_model
import sklearn.neighbors
import sklearn.tree
from sklearn.model_selection import cross_val_score
from sklearn.metrics import mean_squared_error
from sklearn.ensemble import RandomForestRegressor
```

```python
import sklearn.datasets
```

```python
diabetes = sklearn.datasets.load_diabetes()
```

```python
print(diabetes.DESCR)
```

```python
x = diabetes.data
y = diabetes.target
```

```python
print(len(x))
print(len(y))
```

```python
##set up training and testing data to be used throughout
x_train, x_test, y_train, y_test = sklearn.model_selection.train_test_split(
    x, y)
```

```python
##define models for all four types
model1 = sklearn.linear_model.LinearRegression()
model2 = sklearn.neighbors.KNeighborsRegressor(n_neighbors=3)
model3 = sklearn.tree.DecisionTreeRegressor(max_depth=2)
model4 = sklearn.ensemble.RandomForestRegressor(max_depth=2)
```

```python
#perform cross validation on test and train model
#for KNN
loss = cross_val_score(model2,
                       x_train,
                       y_train, 
                       cv=5, 
                       scoring='neg_mean_squared_error')
```

```python
#for KNN
k_range = range(1, 20)
k_scores = []
for k in k_range:
    knn = sklearn.neighbors.KNeighborsRegressor(n_neighbors=k)
    loss = cross_val_score(knn,
                           x_train,
                           y_train, 
                           cv=5, 
                           scoring='neg_mean_squared_error')
    k_scores.append(np.sqrt(-loss).mean())
    
plt.scatter(k_range, k_scores)
plt.xlabel('Value of K for KNN')
plt.ylabel('Cross-Validated MSE')
plt.show()
print(np.mean(np.sqrt(-loss)))
```

The ideal value for K-nearest neighbors would be either 10 or 17, as these two values appear to be the lowest on the graph. For the sake of this problem, I'll redefine the model using KNN = 10.

```python
##redefine model 2 for KNN
model2 = sklearn.neighbors.KNeighborsRegressor(n_neighbors=10)
```

```python
#use for decision tree
loss2 = cross_val_score(model3,
                       x_train,
                       y_train, 
                       cv=5, 
                       scoring='neg_mean_squared_error')
```

```python
#find range of max depth using cross validation
depth_range = range(1, 20)
depth_max = []
for k in depth_range:
    depth = sklearn.tree.DecisionTreeRegressor(max_depth=5)
    loss2 = cross_val_score(depth,
                           x_train,
                           y_train, 
                           cv=5, 
                           scoring='neg_mean_squared_error')
    depth_max.append(np.sqrt(-loss2).mean())
plt.scatter(depth_range, depth_max)
plt.xlabel('Value of Depth for Decision Tree')
plt.ylabel('Cross-Validated MSE')
plt.show()
    
print(np.mean(np.sqrt(-loss2)))
```

```python
#redefine model for Decision Tree using Max Depth
model3 = sklearn.tree.DecisionTreeRegressor(max_depth=12)
```

```python
#use for random forest
loss3 = cross_val_score(model4,
                       x_train,
                       y_train, 
                       cv=5, 
                       scoring='neg_mean_squared_error')
```

```python
#find range of max depth using cross validation
depth_range1 = range(1, 20)
depth_max1 = []
for k in depth_range1:
    depth1 = sklearn.ensemble.RandomForestRegressor(max_depth=5)
    loss3 = cross_val_score(depth1,
                           x_train,
                           y_train, 
                           cv=5, 
                           scoring='neg_mean_squared_error')
    depth_max1.append(np.sqrt(-loss3).mean())
plt.scatter(depth_range1, depth_max1)
plt.xlabel('Value of Depth for Random Forest')
plt.ylabel('Cross-Validated MSE')
plt.show()
    
print(np.mean(np.sqrt(-loss3)))
```

```python
#redefine model for forest
model4 = sklearn.ensemble.RandomForestRegressor(max_depth=18)
```

```python
#fit all of the models
model1.fit(x_train, y_train)  #fit for LR
model2.fit(x_train, y_train)  #fit for KNN
model3.fit(x_train, y_train)  #fit for decision tree
model4.fit(x_train, y_train)  #fit for random forest
y_pred1 = model1.predict(x_test)
y_pred2 = model2.predict(x_test)
y_pred3 = model3.predict(x_test)
y_pred4 = model4.predict(x_test)
```

```python
#print MSE values for linear regression, k-nearest neighbor, decision tree, random forest
print('MSE_LR = ', mean_squared_error(y_test, y_pred1))
print('MSE_knn = ', mean_squared_error(y_test, y_pred2))
print('MSE_DT = ', mean_squared_error(y_test, y_pred3))
print('MSE_RF = ', mean_squared_error(y_test, y_pred4))
```

```python
import numpy as np
import matplotlib
from matplotlib import pyplot as plt
```

```python
import sklearn.datasets
```

```python
# define data set
diabetes = sklearn.datasets.load_diabetes()
```

```python
print(diabetes.DESCR)
```

```python
x = diabetes.data
y = diabetes.target
```

```python

print(x[0,3])
print(len(x))
print(y[11])
print(len(y))
```

```python
x_axis = x[:,2]
y_axis = y
plt.scatter(x_axis, y_axis)
```

```python
import sklearn.linear_model
```

```python
model = sklearn.linear_model.LinearRegression()
```

```python
#reshape x and train 
x_axis = x_axis.reshape(-1,1)
```

```python
model.fit(x_axis, y_axis)
```

```python
##predict and test the model
x_test = [[25]]
model.predict(x_test)
```

```python
##visualize predictions with y_pred

plt.plot(x_axis,y_axis,'ko')

x_new = x_axis.reshape(-1,1)
y_pred = model.predict(x_new)
plt.plot(x_new, y_pred, 'r')

plt.show()
```

```python
#Parameters for model equation, print coefficients
print(model.coef_, model.intercept_)
```

```python
#Plot model using equation

plt.plot(x_axis,y_axis,'ko')
x_new = x_axis.reshape(-1,1)

#the predicted y values are from a model equation, 
#rather than the results of a function
y_pred = model.intercept_ + model.coef_ * x_new

plt.plot(x_new, y_pred, 'r')
plt.show()
```

```python
from sklearn.metrics import mean_squared_error, r2_score

print('MSE = ', mean_squared_error(y_axis, model.predict(x_new)))
print('R^2 = ', r2_score(y_axis, model.predict(x_new)))
```
